### 2. Exploratory Data Analysis

Investigating the dataset is a crucial step in the initial phases to build a good quality model. It enables a comprehensive grasp of the data, detection of any possible issues, and uncovering of significant patterns. In this section, we will carefully analyze the dataset to establish a strong basis for future modeling and analysis.

```python
# import libraries for data processing
import pandas as pd
import numpy as np

# import libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
%matplotlib inline


# ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

a/ Data Loading and Inspection

```python
data = pd.read_csv("train.csv")
```

```python
# Display the number of rows and columns in the train set
print(data.shape)
```

```python
data.info()
```

```python
# Display the first 5 rows of the dataset
# This is useful to understand the data's structure, column names, and sample values
data.head()
```

```python
# drop column id because our model does not need this information
data = data.drop(columns=['Id'])
```

b/ Data Cleaning and  Preproccessing

```python
# Identify columns with missing values
missing_values = data.isnull().sum()
columns_with_missing_values = missing_values[missing_values > 0]

# Print columns with missing values
columns_with_missing_values
```

```python
# Plot the missing values
cmap = plt.get_cmap('viridis')
colors = cmap(np.linspace(0, 1, len(missing_values)))


plt.figure(figsize=(12,8))
missing_values[missing_values > 0].sort_values(
    ascending=False).plot(kind='bar', color = colors)
plt.title('Missing Values in Each Column')
plt.xlabel('Columns')
plt.ylabel('Number of Missing Values')
plt.xticks(rotation=45)
plt.show()
```

It appears that this dataset has a considerable number of missing values, therefore we need to take care of it properly before the training steps
